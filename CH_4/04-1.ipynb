```python
# self-attention
import numpy as np
import pandas as pd
import keras
from keras import layers
```

```python
B = 1 ;T = 3 ; d = 2

X = np.array([
    [[1,0], [0,1], [1,1]]
])

print(X) ; print("shape:", X.shape)
```

```python
query = X
key = X
value = X

print("Q:", query)
print("shape:", query.shape)

# key transpose
key_t = np.transpose(key, (0,2,1))
print("key_t:", key_t)
print("shape:", key_t.shape)

# QK^T
score = np.matmul(query, key_t)
print("QK^T:", score)

# softmax
def softmax(x):
    e = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e / np.sum(e, axis=-1, keepdims=True)

S = softmax(score)
print("softmax:", S)

# final output
out = np.matmul(S, value)
print("output:", out)
```

```python
# B : batch size
# T : token 
# embed_dim : embedding vector dimension
# inputs: (B, T, embed_dim)
# att_dim: query, key, value vector dimension, attention hidden dimension

def self_attention(inputs, att_dim):

    print(inputs) ; print(att_dim)

    # Dense : Input vector Linear transformation?  
    # XW+b??
    query = layers.Dense(att_dim)(inputs)
    key = layers.Dense(att_dim)(inputs)
    value = layers.Dense(att_dim)(inputs)

    print(query)
    print(key)
    print(value)

    # axes 0, 1, 2 -> 0,2,1  
    key_t = keras.ops.transpose(key, axes=(0,2,1))
    print(key_t)
    # dot product
    query_key_dot = keras.ops.matmul(query, key_t) / keras.ops.sqrt(att_dim)
    print(query_key_dot)
    # softmax 
    score = keras.activations.softmax(query_key_dot)
    print(score)
    # dot product, final attention score return
    return keras.ops.matmul(score, value)
```
